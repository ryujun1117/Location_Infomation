```python
from matplotlib import pyplot as plt
from matplotlib import patches

patch = patches.Arrow(x=0.25, y=0.25, dx=0.5, dy=0.5, width=0.3, color="black")

fig, ax = plt.subplots(figsize=(6, 6))
ax.add_patch(patch)
ax.axis('off')

# plt.show()
plt.savefig("a1.png")
```
